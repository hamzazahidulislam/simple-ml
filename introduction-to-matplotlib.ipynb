## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (Width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
       ylabel="y-axis"
      )

# 5. Save & show (you save the whole figure)
fig.savefig("./images/simple-plot.png")
```

## Making figures with NumPy arrays

we want:
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0,10,100)
# x[:10]
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
plt.show()
```

```python
5**2
```

```python
# Create a scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
plt.show()
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
plt.show()
```

```python
# Make a plot form dictionary 
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                     "Cashew butter": 12 
                    }
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
ax.set(title="Dan's Nut Butter shop",
      ylabel="Price ($) "
      );
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
plt.show()
```

```python
# Make some data for histograms and plot it.
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5)
                                            )
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
ax4.hist(np.random.random(10))
plt.show()
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5) )
# Plot to each different index
ax[0, 0].plot(x,x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
ax[1, 1].plot(x,x/2)
plt.show()
```
