## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (Width, height)

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
       ylabel="y-axis"
      )

# 5. Save & show (you save the whole figure)
fig.savefig("./images/simple-plot.png")
```

```python
# Make a plot form dictionary 
nut_butter_prices = {"Almond butter": 10,
                    "Peanut butter": 8,
                     "Cashew butter": 12 
                    }
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
ax.set(title="Dan's Nut Butter shop",
      ylabel="Price ($) "
      );
plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
plt.show()
```

```python
# Make some data for histograms and plot it.
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
plt.show()
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5)
                                            )
# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
ax4.hist(np.random.random(10))
plt.show()
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5) )
# Plot to each different index
ax[0, 0].plot(x,x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x,y)
ax[1, 1].hist(np.random.random(10))
plt.show()
```

### Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method.
* When plotting something more advanced, use the OO method.

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
# len(over_50)
```

```python
## Pyplot method
over_50.plot(kind="scatter",
            x="age",
             y="chol",
             c="target",
             figsize=(10,6)
            )
plt.show()
```

```python
over_50
```

```python
## OO method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
             y="chol",
             c="target",
             ax=ax
            )
# ax.set_xlim([45,100])
plt.show()
```
